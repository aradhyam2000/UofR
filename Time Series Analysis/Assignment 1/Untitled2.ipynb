```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
travel = pd.read_csv (r'BTS_Air_Rail_Vehicle_Miles.csv')
```

```python
travel.Month = pd.to_datetime(travel.Month,  format="%b-%y")
```

```python
sns.lineplot(x = "Month", y = "Air" , data =travel )
from matplotlib import rcParams
plt.rcParams["figure.figsize"] = (50, 20)
sns.set(font_scale=4)
plt.show()
```

```python
#Trend
air= travel[["Air"]] 
air.rolling(5).mean().plot(figsize=(12,5), linewidth=2, fontsize=14)
plt.xlabel('Month')
```
