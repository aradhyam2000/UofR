```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
travel = pd.read_csv (r'BTS_Air_Rail_Vehicle_Miles.csv')
travel.head()
```

Changing format for Month column from Jan-90 to 1990-01-01

```python
travel.Month = pd.to_datetime(travel.Month,  format="%b-%y")
travel.head()
```

Air Travel

```python
#Time Series Plot for Air Travel
plt.figure(figsize=(16, 3))
plt.plot(travel.Month, travel.Air)
plt.xlabel('Month (t)')
plt.ylabel('Air (yt)')
plt.title("Time Series Plot for Air Travel")
plt.plot(travel.Month[0], travel.Air[0], marker="x", markersize=5, markeredgecolor="red")
plt.plot(travel.Month[1], travel.Air[1], marker="x", markersize=5, markeredgecolor="blue")
plt.plot(travel.Month[2], travel.Air[2], marker="x", markersize=5, markeredgecolor="yellow")
plt.plot(travel.Month[7], travel.Air[7], marker="o", markersize=7, markeredgecolor="pink")
plt.show()
```

```python
#t1, t2 and t3
travel.Month[0] , travel.Month[1] , travel.Month[2]
```

```python
# Mean for Air travel
amean = statistics.mean(travel.Air)
amean
```

FORECASTING SHAMPOO SALES

```python
#Reading shampoo sales data
shampoo = pd.read_csv (r'ShampooSales.csv')
shampoo.head()
```

```python
# Changing Month format
shampoo.Month = pd.to_datetime(shampoo.Month,  format="%b-%y")
# Changing Column name from Shampoo Sales to Shampoo_Sales
shampoo.columns = shampoo.columns.str.replace(' ', '_')
shampoo.head()
```

```python
#Time Series Plot for Shampoo Sales
plt.figure(figsize=(16, 3))
plt.plot(shampoo.Month, shampoo.Shampoo_Sales)
plt.xlabel('Month')
plt.ylabel('Shampoo Sales')
plt.title("Time Series Plot for Shampoo Sales")
plt.show()
```

US beverage product shipments data

```python
# Read US Beverage product shipment data 
ship = pd.read_csv (r'Beverages_Shipment_2020.csv')
ship.head()
```

```python
#Changing Month format
ship.Month = pd.to_datetime(ship.Month,  format="%b-%y")
#Changing column name
ship.columns = ['Month','Dollars']
ship.head()
```

```python
#Time Series Plot for US beverage product shipments data
plt.figure(figsize=(16, 3))
plt.plot(ship.Month, ship.Dollars)
plt.title("Time Series Plot for US beverage product shipment")
plt.xlabel('Month')
plt.ylabel('Dollars (in Millions)')
plt.plot(ship.Month[5], ship.Dollars[5], marker="o", markersize=7, markeredgecolor="black")
plt.show()
```

```python
# Autocorrelation Function Plot for US beverage product shipments data
plt.figure(figsize=(16, 3))
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(ship['Dollars'], lags=30)
plt.show()
```

Data on US coal production 

```python
# Reading Data on US coal production
coal = pd.read_csv (r'Coal_Production_US_2020.csv')
coal.head()
```

```python
# Changing column name
coal.columns = ['Year','Production']
coal.head()
```

```python
# Time series plot for Data on US coal production
plt.figure(figsize=(16, 3))
plt.plot(coal.Year, coal.Production)
plt.xlabel('Year')
plt.title("Time Series Plot for US coal production")
plt.ylabel('Production')
plt.show()
```

```python
# Autocorrelation function plot for Data on US coal production
plot_acf(coal['Production'], lags=30)
plt.show()
```

```python
# Time series plot for Data for US coal production after first differencing
from statsmodels.tsa.statespace.tools import diff
plt.figure(figsize=(16, 3))
ts_diff = diff(coal.iloc[:,1])
plt.xlabel('Year')
plt.title("Time Series Plot for US coal production after first differencing")
plt.ylabel('Production')
plt.plot(ts_diff)
plt.show()
```

```python
# Autocorrelation function plot for Data on US coal production after first differencing
plot_acf(ts_diff, lags=30)
plt.show()
```

For Self Learning

```python
#Trend for Air travel
air= travel[["Air"]] 
air.rolling(5).mean().plot(figsize=(16,3), linewidth=2, fontsize=14)
plt.xlabel('Month')
```

```python
#Seasonality for Air travel

air.diff().plot(figsize=(16,3), linewidth=2, fontsize=14)
plt.xlabel('Month')
```

```python
#Trend for Rail Travel
rail = travel[["Rail"]] 
rail.rolling(5).mean().plot(figsize=(16,3), linewidth=2, fontsize=14)
plt.xlabel('Month')
```

```python
#Seasonality for Rail Travel

rail.diff().plot(figsize=(16,3), linewidth=2, fontsize=14)
plt.xlabel('Month')
```

```python
#Trend for Vehicle Travel
vehicle = travel[["Vehicle"]] 
vehicle.rolling(5).mean().plot(figsize=(16,3), linewidth=2, fontsize=14)
plt.xlabel('Month')
```

```python
#Seasonality for Vehicle Travel

vehicle.diff().plot(figsize=(16,3), linewidth=2, fontsize=14)
plt.xlabel('Month')
```
