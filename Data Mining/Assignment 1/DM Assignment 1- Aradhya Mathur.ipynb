     Applying Transformation 

Log Transformation

```python
from collections import Counter
array_homa = np.array(homa)  
homa_reshape = array_homa.reshape((-1))
homa_reshape
```

```python
from sklearn import preprocessing
a_homa = np.array(homa_reshape)
a_log = np.log([a_homa])
print(a_log)
```

```python
homalog = pd.DataFrame(a_log)
log_homa = homalog.T
```

```python
plt.hist(log_homa, 8, color='bisque', alpha=1, rwidth=0.9, edgecolor="black", linewidth=1)
plt.title('Histogram after Log Transformation')
plt.xlabel('Log Homa')
plt.ylabel('Count')
plt.show()
#Histogram after Log Transformation
```

```python
log_homa.skew(axis = 0, skipna = True) #Skewness
```

Reduced skewness of HOMA from 3.812 to 0.914 by applying log transformation.

SQRT Transformation

```python
a_sqrt = np.sqrt([a_homa])
print(a_sqrt)
```

```python
homasqrt = pd.DataFrame(a_sqrt)
sqrt_homa = homasqrt.T
```

```python
plt.hist(sqrt_homa, 8, color='bisque', alpha=1, rwidth=0.9, edgecolor="black", linewidth=1)
plt.title('Histogram after SQRT Transformation')
plt.xlabel('SQRT Homa')
plt.ylabel('Count')
plt.show()
#Histogram after SQRT Transformation
```

```python
sqrt_homa.skew(axis = 0, skipna = True) # Skewness
```

Reduced skewness of HOMA from 3.812 to 2.176 by applying sqrt transformation.


CBRT Transformation

```python
a_cbrt = np.cbrt([a_homa])
print(a_cbrt)
```

```python
homacbrt = pd.DataFrame(a_cbrt)
cbrt_homa = homacbrt.T
plt.hist(cbrt_homa, 8, color='bisque', alpha=1, rwidth=0.9, edgecolor="black", linewidth=1)
plt.title('Histogram after CBRT Transformation')
plt.xlabel('CBRT Homa')
plt.ylabel('Count')
plt.show()
#Histogram after CBRT Transformation
```

```python
cbrt_homa.skew(axis = 0, skipna = True)  # finding skewness
```

Reduced skewness of HOMA from 3.812 to 1.699 by applying cbrt transformation.

Comparing log, sqrt and cbrt transformation, we can infer that log transform works best followed by cbrt and sqrt respectively.
